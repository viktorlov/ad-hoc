```python
import json
from warnings import filterwarnings as _fw_

import pandas as pd

_fw_('ignore')
```

```python
df = pd.read_csv('df.csv', encoding='utf-8').drop(columns='Unnamed: 0')
# df
```

```python
with open('binary_columns.json', 'r', encoding='utf-8') as file:
    binary_columns = json.load(file)
# binary_columns[:-2]
```

```python
target = binary_columns[0:13]
school_head5 = binary_columns[14:30]
school_tail5 = binary_columns[31:46]
course = binary_columns[47:56]
profession = binary_columns[57:66]
```

```python
df_P = df.copy()
```

```python
df_P["TARGET ALL"] = (df_P[target].sum(axis=1) / len(target) * 100).apply(int) // 10
df_P["SCHOOL_HEAD5 ALL"] = (df_P[school_head5].sum(axis=1) / len(school_head5) * 100).apply(int) // 10
df_P["SCHOOL_TAIL5 ALL"] = (df_P[school_tail5].sum(axis=1) / len(school_tail5) * 100).apply(int) // 10
df_P["COURSE ALL"] = (df_P[course].sum(axis=1) / len(course) * 100).apply(int) // 10
df_P["PROFESSION ALL"] = (df_P[profession].sum(axis=1) / len(profession) * 100).apply(int) // 10
df_P
```

```python
last_8 = df_P.columns.tolist()[-8:]
last_8.insert(0, "RESPONSE_ID")
df_last_8 = df_P[last_8]
df_last_8
```

```python
df_last_8['_AGE'] = df_last_8['_AGE'].fillna(31.0)
df_last_8['AGE_CATEGORY'] = (1 + df_last_8['_AGE'] // 10).apply(int) * 10
df_last_8
```
